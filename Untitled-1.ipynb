```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.datasets import ImageFolder
from torchvision import transforms, models
from torch.utils.data import DataLoader

# ---------------------------
# DEVICE
# ---------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using:", device)

# ---------------------------
# TRANSFORMS
# ---------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# ---------------------------
# PATHS
# ---------------------------
train_path = r"C:\Users\Mahesh\Desktop\Strawberry leaf disease 2\archive (1)\train"
val_path   = r"C:\Users\Mahesh\Desktop\Strawberry leaf disease 2\archive (1)\val"
test_path  = r"C:\Users\Mahesh\Desktop\Strawberry leaf disease 2\archive (1)\test"

# ---------------------------
# LOAD DATASETS
# ---------------------------
train_set = ImageFolder(train_path, transform=transform)
val_set   = ImageFolder(val_path, transform=transform)
test_set  = ImageFolder(test_path, transform=transform)

train_loader = DataLoader(train_set, batch_size=8, shuffle=True)
val_loader   = DataLoader(val_set, batch_size=8, shuffle=False)
test_loader  = DataLoader(test_set, batch_size=8, shuffle=False)

class_names = train_set.classes
num_classes = len(class_names)
print("Classes:", class_names)

# ---------------------------
# MODEL (ResNet18 Transfer Learning)
# ---------------------------
model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(device)

# ---------------------------
# LOSS & OPTIMIZER
# ---------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# ---------------------------
# TRAINING FUNCTION
# ---------------------------
def train_model():
    model.train()
    for epoch in range(10):    # Train 10 epochs
        running_loss = 0.0
        correct = 0
        total = 0

        for imgs, labels in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        print(f"Epoch {epoch+1}/10 - Loss: {running_loss:.4f}, Accuracy: {100*correct/total:.2f}%")

# ---------------------------
# VALIDATION FUNCTION
# ---------------------------
def evaluate(loader, name="Validation"):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for imgs, labels in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    print(f"{name} Accuracy: {100*correct/total:.2f}%")

# ---------------------------
# RUN TRAINING
# ---------------------------
train_model()
evaluate(val_loader, "Validation")
evaluate(test_loader, "Test")

# ---------------------------
# SAVE MODEL
# ---------------------------
torch.save(model.state_dict(), "strawberry_model.pth")
print("Model saved as strawberry_model.pth")
```

```python
import torch
import torch.nn as nn
from torchvision import transforms, models
from PIL import Image
import torch.nn.functional as F
import os

# -------------------------------------------
# DEVICE
# -------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -------------------------------------------
# CLASS NAMES (same as training)
# -------------------------------------------
class_names = [
    'angular_leafspot',
    'anthracnose_fruit_rot',
    'blossom_blight',
    'gray_mold',
    'leaf_spot',
    'powdery_mildew_fruit',
    'powdery_mildew_leaf'
]

# -------------------------------------------
# TRANSFORMS
# -------------------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# -------------------------------------------
# LOAD TRAINED CNN MODEL
# -------------------------------------------
model = models.resnet18()
model.fc = nn.Linear(model.fc.in_features, len(class_names))
model.load_state_dict(torch.load("strawberry_model.pth", map_location=device))
model.to(device)
model.eval()

# -------------------------------------------
# FINAL PREDICTION FUNCTION
# -------------------------------------------
def predict_leaf_disease(image_path, threshold=0.50):
    """
    Final prediction function (CNN only)
    """

    # Load image
    img = Image.open(image_path).convert("RGB")
    img_tensor = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(img_tensor)
        probs = F.softmax(outputs, dim=1)[0]

        pred_idx = torch.argmax(probs).item()
        confidence = probs[pred_idx].item()

    # -----------------------
    # CLEAN INTERFACE OUTPUT
    # -----------------------
    print("\n===============================")
    print("      🍓 LEAF DISEASE RESULT")
    print("===============================\n")

    print(f"🖼️ Image: {image_path}")
    print(f"🔍 Predicted Class: {class_names[pred_idx]}")
    print(f"📊 Confidence: {confidence:.2f}")
    print(f"🔢 Probabilities: {probs.cpu().tolist()}")

    if confidence < threshold:
        print("\n⚠️ WARNING: Model is not confident.")
        print("   This image may be unknown or unclear.\n")

    print("===============================")


# -------------------------------------------
# RUN PREDICTION
# -------------------------------------------
image_path = r"C:\Users\Mahesh\Desktop\New folder\al1.jpg"

if os.path.exists(image_path):
    predict_leaf_disease(image_path)
else:
    print("❌ Image not found:", image_path)
```
